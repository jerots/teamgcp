```python
import numpy as np
import pandas as pd

import glob
import re

# user='jeremyong'
txt='behminyan'
```

```python
users=[]
folders=glob.glob("data/*")

for fd in folders:
    users.append(fd[fd.rfind('/')+1:])
    
print 'USERNAMES:', users
```

```python
input = pd.read_csv('data/jeremyong/behminyan_1.csv')
input
```

```python
def extractDurations(table):
    letters=table.letter
    DTs=table.down_time
    UTs=table.up_time

    HDlist=[]
    UDDlist=[]
    DDDlist=[]

    for i in range(len(input)):
        holddur=UTs[i]-DTs[i]
        HDlist.append(holddur)

        if(i< len(input)-1):
            updowndur=DTs[i+1]-UTs[i]
            UDDlist.append(updowndur)
#             print holddur+updowndur
            DDDlist.append(holddur+updowndur)
        else:
            UDDlist.append(0)
            DDDlist.append(0)

    return letters, HDlist, UDDlist, DDDlist
```

```python
df_rows = []
cols=["text","user"]
letters=""

for user in users:
    filenames=glob.glob("data/"+user+"/"+txt+"*.csv")
    print 'FILENAMES:', filenames
    
    for filename in filenames:

        input = pd.read_csv(filename)
        letters, HDlist, UDDlist, DDDlist = extractDurations(input)
    #     username = re.match(r"(.+)_(.+).csv", filename).groups()[1]

        userrow={}
        userrow['text'] = txt
        userrow['user'] = user

        for i in range(len(letters)):
            if(letters[i] == 'return'):
                break
            userrow['hd'+str(i)+'_'+letters[i]]=HDlist[i]
            userrow['udd'+str(i)+'_'+letters[i]]=UDDlist[i]
            userrow['ddd'+str(i)+'_'+letters[i]]=DDDlist[i]
        df_rows.append(userrow)
    
for i in range(len(letters)): # SOON TO CHANGE
        if(letters[i] == 'return'):
            break
        cols.append('hd'+str(i)+'_'+letters[i])
        cols.append('udd'+str(i)+'_'+letters[i])
        cols.append('ddd'+str(i)+'_'+letters[i])
```

```python
combined = pd.DataFrame(df_rows,columns=cols)
combined
```

```python
# Taken from: https://www.peterbe.com/plog/uniqifiers-benchmark
def unique (seq, idfun=None): # order preserving
    if idfun is None:
        def idfun(x): 
            return x
    seen = {}
    result = []
    for item in seq:
        marker = idfun(item)
        if marker in seen: 
            continue
        seen[marker] = 1
        result.append(item)
    return result
```

```python
cats=combined['user'].tolist()
catlist=unique(cats)
print catlist

combined['cat'] = [catlist.index(x) for x in combined.user]
combined
```

# Recursive Feature Elimination

```python
traindata=combined.drop(['text','user','cat'], axis=1).as_matrix(columns=None)
labels=combined['cat'].tolist()
```

```python
# Recursive Feature Elimination
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# create a base classifier used to evaluate a subset of attributes
model = LogisticRegression()
# create the RFE model and select 3 attributes
rfe = RFE(model, 3)
# rfe = rfe.fit(dataset.data, dataset.target)
rfe = rfe.fit(traindata, labels)
# summarize the selection of the attributes
print(rfe.support_)
print(rfe.ranking_)
```

```python
train_features=['hd0_b', 'udd0_b', 'ddd0_b', 'hd1_e', 'udd1_e', 'ddd1_e', 'hd2_h', 'udd2_h', 'ddd2_h', 'hd3_space', 'udd3_space', 'ddd3_space', 'hd4_m', 'udd4_m', 'ddd4_m', 'hd5_i', 'udd5_i', 'ddd5_i', 'hd6_n', 'udd6_n', 'ddd6_n', 'hd7_space', 'udd7_space', 'ddd7_space', 'hd8_y', 'udd8_y', 'ddd8_y', 'hd9_a', 'udd9_a', 'ddd9_a', 'hd10_n', 'udd10_n', 'ddd10_n']
```

```python
res1=pd.DataFrame({'feature': train_features, 'importance':rfe.ranking_}).sort_values(by=['importance'],ascending=True)
res1
```

```python
# Feature Importance
# from sklearn import datasets
from sklearn import metrics
from sklearn.ensemble import ExtraTreesClassifier

# fit an Extra Trees model to the data
model = ExtraTreesClassifier()
model.fit(traindata, labels)
# display the relative importance of each attribute
print(model.feature_importances_)
```

```python
res=pd.DataFrame({'feature':train_features, 'importance':model.feature_importances_}).sort_values(by=['importance'],ascending=False)
res
```
