```python
def extractDurations(table):
    letters=table.letter
    DTs=table.down_time
    UTs=table.up_time

    HDlist=[]
    UDDlist=[]
    DDDlist=[]

    for i in range(len(input)):
        holddur=UTs[i]-DTs[i]
        HDlist.append(holddur)

        if(i< len(input)-1):
            updowndur=DTs[i+1]-UTs[i]
            UDDlist.append(updowndur)
#             print holddur+updowndur
            DDDlist.append(holddur+updowndur)
        else:
            UDDlist.append(0)
            DDDlist.append(0)

    return letters, HDlist, UDDlist, DDDlist
```

```python
# Taken from: https://www.peterbe.com/plog/uniqifiers-benchmark
def unique (seq, idfun=None): # order preserving
    if idfun is None:
        def idfun(x): 
            return x
    seen = {}
    result = []
    for item in seq:
        marker = idfun(item)
        if marker in seen: 
            continue
        seen[marker] = 1
        result.append(item)
    return result
```

```python
shortlist=['udd0_b', 'udd2_h', 'udd4_m', 'udd10_n', 'ddd5_i', 'udd1_e', 'ddd0_b', 'hd9_a', 'ddd3_space', 'udd5_i']
shortlist.append('user')
dataset=combined[shortlist].drop('user', axis=1).as_matrix(columns=None)
labels=combined['user']
```

```python
def separateByClass(dataset,labels):
    separated = {}
    for i in range(len(dataset)):
        vector = dataset[i] # vector = row of data
        label = labels[i]
        if (label not in separated):
            separated[label] = []
        separated[label].append(vector)
    return separated
```

```python
sep=separateByClass(dataset,labels)
print sep
```

```python
import math
def mean(numbers):
    return sum(numbers)/float(len(numbers))
 
def stdev(numbers):
    avg = mean(numbers)
    variance = sum([pow(x-avg,2) for x in numbers])/float(len(numbers)-1)
    return math.sqrt(variance)

# numbers = [1,2,3,4,5]
# print('Summary of {0}: mean={1}, stdev={2}').format(numbers, mean(numbers), stdev(numbers))
```

```python
def summarize(dataset):
    summaries = [(mean(attribute), stdev(attribute)) for attribute in zip(*dataset)]
    return summaries
```

```python
summary=summarize(dataset)
print('Attribute summaries: {0}').format(summary)
```

```python
def summarizeByClass(dataset,labels):
    separated = separateByClass(dataset,labels)
    summaries = {}
    for classValue, instances in separated.iteritems():
        summaries[classValue] = summarize(instances)
    return summaries
```

```python
summary = summarizeByClass(dataset,labels)
print('Summary by class value: {0}').format(summary)
```

```python
def calculateProbability(x, mean, stdev):
    exponent = math.exp(-(math.pow(x-mean,2)/(2*math.pow(stdev,2))))
    return (1 / (math.sqrt(2*math.pi) * stdev)) * exponent
```

```python
x = 71.5
mean = 73
stdev = 6.2
probability = calculateProbability(x, mean, stdev)
print('Probability of belonging to this class: {0}').format(probability)
```

```python
def calculateClassProbabilities(summaries, inputVector):
    probabilities = {}
    for classValue, classSummaries in summaries.iteritems():
        probabilities[classValue] = 1
        for i in range(len(classSummaries)):
            mean, stdev = classSummaries[i]
            x = inputVector[i]
            probabilities[classValue] *= calculateProbability(x, mean, stdev)
    return probabilities
```

```python
summaries = {0:[(1, 0.5)], 1:[(20, 5.0)]}
inputVector = [1.1, '?']
probabilities = calculateClassProbabilities(summaries, inputVector)
print('Probabilities for each class: {0}').format(probabilities)
```

```python
def predict(summaries, inputVector):
    probabilities = calculateClassProbabilities(summaries, inputVector)
    bestLabel, bestProb = None, -1
    for classValue, probability in probabilities.iteritems():
        if bestLabel is None or probability > bestProb:
            bestProb = probability
            bestLabel = classValue
    return bestLabel
```

```python
summaries = {'A':[(1, 0.5)], 'B':[(20, 5.0)]}
inputVector = [1.1, '?']
result = predict(summaries, inputVector)
print('Prediction: {0}').format(result)
```
